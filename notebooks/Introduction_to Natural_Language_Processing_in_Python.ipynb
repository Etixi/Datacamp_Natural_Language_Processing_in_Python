###**2. Qu'est-ce que la tokenisation ?**

+ ***`La tokenisation est le processus de transformation d'une chaîne ou d'un document en morceaux plus petits, que nous appelons des jetons`. Il s'agit généralement d'une étape dans le processus de préparation d'un texte pour le traitement du langage naturel. Il existe de nombreuses théories et règles différentes concernant la tokenisation, et vous pouvez créer vos propres règles de tokenisation à l'aide d'expressions régulières, mais normalement la tokenisation fera des choses comme séparer des mots ou des phrases, souvent séparer la ponctuation ou vous pouvez même simplement tokeniser des parties d'une chaîne comme séparer tous les hashtags dans un Tweet.***

####**3. bibliothèque nltk**


+ ***Une bibliothèque couramment utilisée pour la tokenisation simple est `nltk`, la bibliothèque de la boîte à outils en langage naturel. Voici un court exemple d'utilisation de la méthode `word_tokenize` pour décomposer une chaîne en jetons. Nous pouvons voir d'après le résultat que les mots sont séparés et que la ponctuation est également des jetons individuels.***

####**4. Pourquoi tokeniser ?**

+ ***Pourquoi s'embêter avec la tokenisation ? Parce qu'il peut nous aider dans certaines tâches simples de traitement de texte, telles que le mappage d'une partie du discours, la correspondance de mots courants et peut-être la suppression de jetons indésirables tels que des mots courants ou des mots répétés. Ici, nous avons un bon exemple. La phrase est : Je n'aime pas les chaussures de Sam. Lorsque nous le symbolisons, nous pouvons clairement voir la négation dans le pas et nous pouvons voir la possession avec le 's. Ces indicateurs peuvent nous aider à déterminer le sens d'un texte simple.***

####**5. Autres tokenizers nltk**

+ ***Au-delà de la simple tokenisation des mots, NLTK propose de nombreux autres tokenizers que vous pouvez utiliser, y compris ceux avec lesquels vous travaillerez dans ce chapitre. La fonction `sent_tokenize divisera un document en phrases individuelles. Le regexp_tokenize utilise des expressions régulières pour tokeniser la chaîne, vous donnant un contrôle plus granulaire sur le processus. Et le tweettokenizer fait des choses intéressantes comme reconnaître les hashtags, les mentions et quand vous avez trop de symboles de ponctuation après une phrase`. Comme c'est pratique!!!***

####**6. Plus de pratique des regex**

+ ***Vous utiliserez également plus d'expressions régulières dans cette section, non seulement lorsque vous segmentez, mais aussi pour comprendre comment analyser les jetons et le texte. L'utilisation de `re.match et re.search du module regex sont des outils assez essentiels pour le traitement des chaînes Python`. Apprendre quand utiliser la recherche par rapport à la correspondance peut être difficile, alors examinons en quoi ils sont différents. Lorsque nous utilisons la recherche et la correspondance avec le même modèle et la même chaîne avec le modèle au début de la chaîne, nous constatons que nous trouvons des correspondances identiques. C'est le cas de la correspondance et de la recherche de abcde avec le motif abc. Lorsque nous utilisons la recherche d'un motif qui apparaît plus tard dans la chaîne, nous obtenons un résultat, mais nous n'obtenons pas le même résultat en utilisant match. En effet, match essaiera de faire correspondre une chaîne depuis le début jusqu'à ce qu'elle ne puisse plus correspondre. La recherche parcourra la chaîne ENTIÈRE pour rechercher les options de correspondance. Si vous avez besoin de trouver un modèle qui ne se trouve peut-être pas au début de la chaîne, vous devez utiliser la recherche. Si vous voulez être précis sur la composition de la chaîne entière, ou au moins sur le modèle initial, vous devez utiliser match.***

####**2. Qu'est-ce que tf-idf ?**

+ ***`Tf-idf signifie term-frequncy - fréquence de document inverse. Il s'agit d'un modèle de traitement du langage naturel couramment utilisé qui vous aide à déterminer les mots les plus importants dans chaque document du corpus`. L'idée derrière tf-idf est que chaque corpus pourrait avoir plus de mots partagés que de simples mots vides. Ces mots courants sont comme des mots vides et doivent être supprimés ou au moins réduits en importance. Par exemple, si je suis astronome, le ciel peut être utilisé souvent mais n'est pas important, donc je veux minimiser ce mot. TF-Idf fait précisément cela. Il faudra des textes qui partagent un langage commun et s'assurer que les mots les plus courants dans l'ensemble du corpus n'apparaissent pas comme des mots-clés. Tf-idf permet de conserver une pondération élevée pour les mots fréquents spécifiques au document et une pondération faible pour les mots courants de l'ensemble du corpus.***

####**3. Formule Tf-idf**

+ ***L'équation pour calculer les poids peut être décrite comme ceci : Le poids du jeton i dans le document j est calculé en prenant le terme fréquence (ou combien de fois le jeton apparaît dans le document) multiplié par le log du nombre total de documents divisés par le nombre de documents contenant le même terme. Déballons un peu ça. Premièrement, le poids sera faible si le terme n'apparaît pas souvent dans le document car la variable tf sera alors faible. Cependant, le poids sera également faible si le logarithme est proche de zéro, ce qui signifie que l'équation interne est faible. Ici, nous pouvons voir si le nombre total de documents divisé par le nombre de documents contenant le terme est proche de un, alors notre logarithme sera proche de zéro. Ainsi, les mots qui apparaissent dans de nombreux documents ou dans tous les documents auront un poids tf-idf très faible. Au contraire,***
