####**2. L’objet nlp**
+ Au centre de spaCy se trouve l’objet contenant le pipeline de traitement. 

+ Nous appelons généralement cette variable « nlp ». 

+ Par exemple, pour créer un objet nlp anglais, vous pouvez importer la classe de langue anglaise à partir de spacy dot lang dot en et l’instancier. 

+ Vous pouvez utiliser l’objet nlp comme une fonction pour analyser du texte. 

+ Il contient tous les différents composants du pipeline. 

+ Il comprend également des règles spécifiques à la langue utilisées pour tokeniser le texte en mots et en ponctuation.

```python
#The nlp object
# Importation de la classe de langue anglaise
from spacy.lang.en import English
# Création de l'objet nlp
nlp = English()
# contient la ligne de traitement
#inclut les règles spécifiques à la langue pour la tokénisation, etc.
```

####**3. L’objet Doc**
+ Lorsque vous traitez un texte avec l’objet nlp, **spaCy crée un objet Doc, abréviation de « document »**. 

+ **Le Document vous permet d’accéder aux informations sur le texte de manière structurée, et aucune information n’est perdue**. 
+ Le Doc se comporte d’ailleurs comme une séquence Python normale et vous permet d’itérer sur ses jetons, ou d’obtenir un jeton par son index. Mais nous y reviendrons plus tard!





```python
# Created by processing a string of text with the nlp object
doc = nlp("Hello world!")
# Iterate over tokens in a Doc
for token in doc:    
  print(token.text)
```

####**4. L’objet Token**

+ Les objets de jeton représentent les jetons d’un document – par exemple, 
  + un mot 
  + ou un caractère de ponctuation. 
  
+ Pour obtenir un jeton à une position spécifique, vous pouvez indexer dans le Document. 

+ **Les objets de jeton fournissent également divers attributs qui vous permettent d’accéder à plus d’informations sur les jetons**. 

+ Par exemple, l’attribut dot text renvoie le texte du jeton textuel.

```python
doc = nlp("Hello world!")
# Index into the Doc to get a single Token
token = doc[1]
# Get the token text via the .text attribute
print(token.text)
```

####**5. L’objet Span**

+ **Un objet Span est une tranche du document composée d’un ou plusieurs jetons**. 

+ **Il ne s’agit que d’une vue du Doc et ne contient aucune donnée elle-même**. 

+ **Pour créer un Span, vous pouvez utiliser la notation de tranche de Python**. 

+ Par exemple, 1 deux-points 3 créera une tranche à partir du jeton à la position 1, jusqu’à – mais sans inclure! – le jeton à la position 3.



```python
doc = nlp("Hello world!")
# A slice from the Doc is a Span object
span = doc[1:4]
# Get the span text via the .text attribute
print(span.text)
```

####**6. Attributs lexicaux**

+ Ici, vous pouvez voir certains des attributs de jeton disponibles: « i » est l’index du jeton dans le document parent. « text » renvoie le texte du jeton. 

+ **« is alpha », « is punct » et « like num »** renvoient des valeurs booléennes indiquant si le jeton est constitué de caractères alphanumériques, s’il s’agit d’une ponctuation ou s’il ressemble à un nombre. 

+ Par exemple, un jeton « 10 » – un, zéro – ou le mot « dix » – T, E, N. Ces attributs sont également appelés attributs lexicaux : ils font référence à l’entrée dans le vocabulaire et ne dépendent pas du contexte du jeton.



```python
doc = nlp("It costs $5.")
print('Index:   ', [token.i for token in doc])
print("-----"*20)
print('Text:    ', [token.text for token in doc])
print("-----"*20)
print('is_alpha:', [token.is_alpha for token in doc])
print("-----"*20)
print('is_punct:', [token.is_punct for token in doc])
print("-----"*20)
print('like_num:', [token.like_num for token in doc])
```

```python
# Process the text
doc = nlp("In 1990, more than 60% of people in East Asia were in extreme poverty. Now less than 4% are.")

# Iterate over the tokens in the doc
for token in doc:
  # Check if the token resembles a number
  if token.like_num:
    # Get the next token in the document
    next_token = doc[token.i + 1]
    # Check if the next token's text equals '%'
    if next_token.text == '%':
      print('Percentage found:', token.text)
```

###**Modèles statistiques**
####**1. Modèles statistiques**
+ ***Ajoutons un peu plus de puissance à l’objet nlp ! Dans cette vidéo, vous découvrirez les modèles statistiques de spaCy.***

####**2. Que sont les modèles statistiques?**
+ Certaines des choses les plus intéressantes que vous pouvez analyser sont spécifiques au contexte: 
  + **par exemple, si un mot est un verbe ou si une étendue de texte est un nom de personne.**
  + **Les modèles statistiques permettent à spaCy de faire des prédictions en contexte.** 
  + **Cela inclut généralement des balises de partie de la parole, des dépendances syntaxiques et des entités nommées.** 
  + **Les modèles sont formés sur de grands ensembles de données d’exemples de textes étiquetés.** 
  + **Ils peuvent être mis à jour avec plus d’exemples pour affiner leurs prédictions - par exemple, pour mieux performer sur vos données spécifiques.**

####**3. Forfaits modèles**
+ **spaCy fournit un certain nombre de packages de modèles pré-entraînés que vous pouvez télécharger.** 

+ **Par exemple, le package `« en_core_web_sm »` est un petit modèle anglais qui prend en charge toutes les fonctionnalités de base et est formé sur le texte Web**. 

+ **La méthode `spacy.load` charge un package de modèle par nom et retourne un objet nlp. Le package fournit les poids binaires qui permettent à spaCy de faire des prédictions. Il inclut également le vocabulaire et les méta-informations pour indiquer à spaCy quelle classe de langue utiliser et comment configurer le pipeline de traitement.**

####**4. Prédiction des balises de partie de la parole**

+ Jetons un coup d’œil aux prédictions du modèle. 

+ Dans cet exemple, nous utilisons spaCy pour prédire les balises de partie de la parole, les types de mots en contexte. 

+ Tout d’abord, nous chargeons le petit modèle anglais et recevons un objet nlp. 

+ Ensuite, nous traitons le texte « Elle a mangé la pizza ». Pour chaque jeton dans le Doc, nous pouvons imprimer le texte et l’attribut `« pos_»`, la balise de partie de discours prévue. 

+ Dans spaCy, les attributs qui renvoient des chaînes se terminent généralement par un trait de soulignement – les attributs sans le trait de soulignement renvoient un ID. 

+ Ici, le modèle a correctement prédit « ate » comme verbe et « pizza » comme nom.

####**5. Prédiction des dépendances syntaxiques**
+ En plus des balises de partie de la parole, nous pouvons également prédire comment les mots sont liés. 
+ Par exemple, si un mot est le sujet de la phrase ou un objet. 
+ L’attribut « dep underscore » renvoie l’étiquette de dépendance prévue. 
+ L’attribut `head` renvoie le jeton `head` syntaxique. Vous pouvez également le considérer comme le jeton parent auquel ce mot est attaché.

####**6. Schéma d’étiquetage des dépendances**
+ Pour décrire les dépendances syntaxiques, spaCy utilise un schéma d’étiquetage standardisé. 
+ Voici un exemple de quelques étiquettes courantes : 
  + **Le pronom « She » est un sujet nominal attaché au verbe - dans ce cas, « manger »**. 
  + **Le substantif « pizza » est un objet direct attaché au verbe « ate »**. 
  + **Il est mangé par le sujet, « elle ». Le déterminant « le », également connu sous le nom d’article, est attaché au nom « pizza ».**

####**7. Prédiction des entités nommées**

+ **Les entités nommées sont des « objets du monde réel » auxquels un nom est attribué – par exemple, une personne, une organisation ou un pays**. 

+ **La propriété `doc.ents` vous permet d’accéder aux entités nommées prédites par le modèle.** 

+ **Il retourne un itérateur d’objets Span, afin que nous puissions imprimer le texte de l’entité et l’étiquette de l’entité à l’aide de l’attribut `« label_»`.** 

+ Dans ce cas, le modèle prédit correctement « Apple » en tant qu’organisation, « Royaume-Uni » en tant qu’entité géopolitique et « 1 milliard de dollars » en argent.

####**8. Astuce : la méthode d’explication**
+ Un petit conseil : pour obtenir des définitions pour les balises et étiquettes les plus courantes, vous pouvez utiliser la fonction d’assistance `spacy.explain`. 

+ Par exemple, « GPE » pour entité géopolitique n’est pas vraiment intuitif – mais `spacy.explain` peut vous dire qu’il fait référence à des pays, des villes et des États. 

+ Il en va de même pour les balises de partie de la parole et les étiquettes de dépendance.

###**Correspondance basée sur des règles**

####**1. Correspondance basée sur des règles**
+ ***Dans cette vidéo, nous allons jeter un coup d’œil au matcher de spaCy, qui vous permet d’écrire des règles pour trouver des mots et des phrases dans le texte.***

####**2. Pourquoi pas seulement des expressions régulières?**

+ **Par rapport aux expressions régulières, le matcher fonctionne avec des objets Doc et Token au lieu de chaînes uniquement.** 

+ **C’est aussi plus flexible : vous pouvez rechercher des textes mais aussi d’autres attributs lexicals**. 

+ **Vous pouvez même écrire des règles qui utilisent les prédictions du modèle. Par exemple, trouvez le mot « canard » uniquement s’il s’agit d’un verbe et non d’un nom.**

####**3. Modèles de correspondance**
+ Les modèles de correspondance sont des listes de dictionnaires. 
+ Chaque dictionnaire décrit un jeton. Les clés sont les noms des attributs de jeton, mappés à leurs valeurs attendues. 

+ Dans cet exemple, nous recherchons deux jetons avec le texte « iPhone » et « $X$ ». 

+ Nous pouvons également faire correspondre sur d’autres attributs de jeton. 

+ Ici, nous recherchons deux jetons dont les formes minuscules sont égales à « iphone » et « $x$ ». 

+ Nous pouvons même écrire des modèles en utilisant des attributs prédits par le modèle. 

+ Ici, nous faisons correspondre un jeton avec le lemme « acheter », plus un nom. 

+ ***Le lemme est la forme de base, donc ce modèle correspondrait à des expressions comme « acheter du lait » ou « acheter des fleurs ».***

####**4. Utilisation du Matcher (1)**
+ Pour utiliser un modèle, nous importons d’abord le matcher à partir de spacy dot matcher. 

+ Nous chargeons également un modèle et créons l’objet nlp. 

+ **Le matcher est initialisé avec le vocabulaire partagé, nlp point vocabulaire**.

####**2. Vocabulaire partagé et magasin de chaînes (1)**

+ **spaCy stocke toutes les données partagées dans un vocabulaire, le Vocab**. 

+ **Cela inclut les mots, mais aussi les schémas d’étiquettes pour les balises et les entités**. 

+ **Pour économiser de la mémoire, toutes les chaînes sont codées en ID de hachage**. 

+ **Si un mot apparaît plus d’une fois, nous n’avons pas besoin de l’enregistrer à chaque fois**. 

+ **Au lieu de cela, spaCy utilise une fonction de hachage pour générer un ID et ne stocke la chaîne qu’une seule fois dans le magasin de chaînes**. 

+ **Le magasin de chaînes est disponible sous forme de chaînes de points de vocabulaire nlp**. 

+ **C’est une table de correspondance qui fonctionne dans les deux sens. Vous pouvez rechercher une chaîne et obtenir son hachage, et rechercher un hachage pour obtenir sa valeur de chaîne***. 

+ **En interne, spaCy ne communique que dans les ID de hachage. Cependant, les ID de hachage ne peuvent pas être inversés. Si un mot n’est pas dans le vocabulaire, il n’y a aucun moyen d’obtenir sa chaîne. C’est pourquoi nous avons toujours besoin de faire circuler le vocabulaire partagé.**

####**3. Vocabulaire partagé et magasin de chaînes (2)**

+ ***Pour obtenir le hachage d’une chaîne, nous pouvons le rechercher dans nlp.vocab.strings**.

####**4. Lexèmes : entrées dans le vocabulaire**

+ **Les lexèmes sont des entrées indépendantes du contexte dans le vocabulaire**. 

+ **Vous pouvez obtenir un lexème en recherchant une chaîne ou un ID de hachage dans le vocabulaire**. 

+ **Les lexèmes exposent les attributs, tout comme les jetons**. 

+ **Ils contiennent des informations indépendantes du contexte sur un mot, comme le texte, ou si le mot se compose de caractères alphanumériques. Les lexèmes n’ont pas de balises de partie de la parole, de dépendances ou d’étiquettes d’entité. Ceux-ci dépendent du contexte.**

```python
class EntityMatcher(object):
    name = "entity_matcher"

    def __init__(self, nlp, terms, label):
        patterns = [nlp.make_doc(text) for text in terms]
        self.matcher = PhraseMatcher(nlp.vocab)
        self.matcher.add(label, None, *patterns)

    def __call__(self, doc):
        matches = self.matcher(doc)
        seen_tokens = set()
        new_entities = []
        entities = doc.ents
        for match_id, start, end in matches:
        #    span = Span(doc, start, end, label=match_id)
        #    doc.ents = list(doc.ents) + [span]
            # check for end - 1 here because boundaries are inclusive
            if start not in seen_tokens and end - 1 not in seen_tokens:
                new_entities.append(Span(doc, start, end, label=match_id))
                entities = [
                    e for e in entities if not (e.start < end and e.end > start)
                ]
                seen_tokens.update(range(start, end))

        doc.ents = tuple(entities) + tuple(new_entities)
        return doc
    
terms = (u"cat", u"dog", u"tree kangaroo", u"giant sea spider", u"Barack Obama")
entity_matcher = EntityMatcher(nlp, terms, "PRESIDENT")
nlp.add_pipe(entity_matcher, after="ner")
print(nlp.pipe_names)  # The components in the pipeline

doc = nlp(u"This is a text about Barack Obama and a tree kangaroo")
print([(ent.text, ent.label_) for ent in doc.ents])
```

Les composants de pipeline personnalisés vous permettent d’ajouter votre propre fonction au pipeline spaCy qui est exécuté lorsque vous appelez l’objet NLP sur un texte, par exemple, pour modifier le Doc et y ajouter des données.***

####**2. Pourquoi des composants personnalisés?**

+ **Une fois que le texte est tokenisé et qu’un objet Doc a été créé, les composants de pipeline sont appliqués dans l’ordre**. 

+ **spaCy prend en charge une gamme de composants intégrés, mais vous permet également de définir les vôtres**.

####**3. Anatomie d’un composant (1)**
+ **Fondamentalement, un composant de pipeline est une fonction ou un appelable qui prend un document, le modifie et le renvoie, afin qu’il puisse être traité par le composant suivant dans le pipeline**. 

+ **Les composants peuvent être ajoutés au pipeline à l’aide de la méthode nlp.add.pipe. La méthode prend au moins un argument : la fonction composante.**

####**4. Anatomie d’un composant (2)**

+ **Pour spécifier `where` pour ajouter le composant dans le pipeline, vous pouvez utiliser les arguments de mots clés suivants :**
  + **Si vous définissez last sur True, le composant est ajouté en dernier dans le pipeline**. 
  + **Il s’agit du comportement par défaut.** 
  + **Si vous définissez d’abord sur True, le composant sera ajouté en premier dans le pipeline, juste après le tokenizer.** 
  + **Les arguments « before » et « after » vous permettent de définir le nom d’un composant existant pour ajouter le nouveau composant avant ou aprè**. 
  + **Par exemple, avant égal à « ner » l’ajoutera avant le module de reconnaissance d’entité nommée. L’autre composant pour ajouter le nouveau composant avant ou après doit exister, cependant, spaCy déclenchera une erreur.**

```python
# Import the PhraseMatcher and initialize it
from spacy.matcher import PhraseMatcher

nlp = spacy.load("en_core_web_sm")
matcher = PhraseMatcher(nlp.vocab)
animal_patterns = ['Golden Retriever', 'cat', 'turtle', 'Rattus norvegicus']
animal_patterns = list(nlp.pipe(animal_patterns))
matcher.add('ANIMAL', None, *animal_patterns)
print(animal_patterns)
```

```python
# Define the custom component
@Language.component("animal_component")
def animal_component(doc):
    # Apply the matcher to the doc
    matches = matcher(doc)
    # Create a Span for each match and assign the label 'ANIMAL'
    spans = [Span(doc, start, end, label='ANIMAL')
             for match_id, start, end in matches]
    # Overwrite the doc.ents with the matched spans
    doc.ents = spans
    return doc
```

```python
# Add the component to the pipeline after the 'ner' component 
nlp.add_pipe("animal_component", after = 'ner')
print(nlp.pipe_names)
```

```python
# Process the text and print the text and label for the doc.ents
doc = nlp("I have a cat and a Golden Retriever")
print([(ent.text, ent.label_) for ent in doc.ents])
```

####**2. Définition d’attributs personnalisés**
+ **Les attributs personnalisés vous permettent d’ajouter des métadonnées à Docs, Tokens et Spans.** 

+ **Les données peuvent être ajoutées une seule fois, ou elles peuvent être calculées dynamiquement.** 

+ **Les attributs personnalisés sont disponibles via la propriété point-underscore.** 

+ **Cela indique clairement qu’ils ont été ajoutés par l’utilisateur et non intégrés à spaCy, comme le texte à points symboliques**. 

+ **Les attributs doivent être enregistrés sur les classes globales Doc, Token et Span que vous pouvez importer à partir de jetons spacy dot**. 

+ Vous avez déjà travaillé avec ceux-ci dans les chapitres précédents. 

+ **Pour enregistrer un attribut personnalisé sur le Doc, le Token ou le Span, vous pouvez utiliser la méthode set extension**. 

+ **Le premier argument est le nom de l’attribut. Les arguments de mots-clés vous permettent de définir comment la valeur doit être calculée. Dans ce cas, il a une valeur par défaut et peut être écrasé.**

####**3. Types d’attributs d’extension**
+ **Il existe trois types d’extensions : les extensions d’attribut, les extensions de propriété et les extensions de méthode.**

####**4. Extensions d’attribut**

+ **Les extensions d’attribut définissent une valeur par défaut qui peut être remplacée.** 
  + Par exemple, un attribut personnalisé « is color » sur le jeton dont la valeur par défaut est False. 
  + Sur les jetons individuels, sa valeur peut être modifiée en l’écrasant - dans ce cas, True pour le jeton « bleu ».

####**5. Prolongations de propriété (1)**
+ Les extensions de propriété fonctionnent comme des propriétés en Python : elles peuvent définir une fonction getter et un setter optionnel. 

+ La fonction getter n’est appelée que lorsque vous récupérez l’attribut. 

+ Cela vous permet de calculer la valeur dynamiquement et même de prendre en compte d’autres attributs personnalisés. 

+ Les fonctions Getter prennent un argument : l’objet, dans ce cas, le jeton. 

+ Dans cet exemple, la fonction renvoie si le texte du jeton figure dans notre liste de couleurs. 

+ Nous pouvons ensuite fournir la fonction via l’argument mot-clé getter lorsque nous enregistrons l’extension. Le jeton « blue » renvoie maintenant True pour « is color ».

#####**6. Prolongations de propriété (2)**
+ Si vous souhaitez définir des attributs d’extension sur un Span, vous devez presque toujours utiliser une extension de propriété avec un getter. 

+ Sinon, vous devrez mettre à jour * chaque plage possible * à la main pour définir toutes les valeurs. 

+ Dans cet exemple, la fonction « get has color » prend la plage et renvoie si le texte de l’un des jetons figure dans la liste des couleurs. 

+ Après avoir traité le document, nous pouvons vérifier différentes tranches du document et la propriété personnalisée « has color » renvoie si la plage contient un jeton de couleur ou non.

####**7. Extensions de méthode**
+ Les extensions de méthode font de l’attribut extension une méthode appelable. 

+ Vous pouvez ensuite lui passer un ou plusieurs arguments et calculer dynamiquement des valeurs d’attribut, par exemple, en fonction d’un certain argument ou d’un certain paramètre. 

+ Dans cet exemple, la fonction method vérifie si le document contient un jeton avec un texte donné. Le premier argument de la méthode est toujours l’objet lui-même – dans ce cas, le Doc. 

+ Il est transmis automatiquement lorsque la méthode est appelée. 

+ Tous les autres arguments de fonction seront des arguments sur l’extension de méthode. Dans ce cas, « texte symbolique ». Ici, la méthode personnalisée « has token » renvoie True pour le mot « blue » et False pour le mot « cloud ».

```python
# Define the getter function
def get_has_number(doc):
    # Return if any of the tokens in the doc return True for token.like_num
    return any(token.like_num for token in doc)

# Register the Doc property extension 'has_number' with the getter get_has_number
Doc.set_extension('has_number', getter = get_has_number, force=True)

# Process the text and check the custom has_number attribute 
doc = nlp("The museum closed for five years in 2012.")
print('has_number:', doc._.has_number)
```

```python
# Define the method
def to_html(span, tag):
    # Wrap the span text in a HTML tag and return it
    return '<{tag}>{text}</{tag}>'.format(tag=tag, text=span.text)

# Register the Span property extension 'to_html' with the method to_html
Span.set_extension('to_html', method = to_html)

# Process the text and call the to_html method on the span with the tag name 'strong'
doc = nlp("Hello world, this is a sentence.")
span = doc[0:2]
print(span._.to_html('strong'))
```

```python
import spacy
from spacy.matcher import Matcher

nlp = spacy.load("en_core_web_sm")

matcher = Matcher(nlp.vocab)

TEXTS = ['How to preorder the iPhone X', 'iPhone X is coming', 'Should I pay $1,000 for the iPhone X?', 'The iPhone 8 reviews are here', 'Your iPhone goes up to 11 today', 'I need a new phone! Any tips?']

# Two tokens whose lowercase forms match 'iphone' and 'x'
pattern1 = [{'LOWER': 'iphone'}, {'LOWER': 'x'}]

# Token whose lowercase form matches 'iphone' and an optional digit
pattern2 = [{'LOWER': 'iphone'}, {'IS_DIGIT': True, 'OP': '?'}]

# Add patterns to the matcher
matcher.add('GADGET', [pattern1, pattern2])
#matcher.add('GADGET', pattern2)
```

```python
# Create a Doc object for each text in TEXTS
for doc in nlp.pipe(TEXTS):
    # Find the matches in the doc
    matches = matcher(doc)
    
    # Get a list of (start, end, label) tuples of matches in the text
    entities = [(start, end, 'GADGET') for match_id, start, end in matches]
    print(doc.text, entities)   
```

```python
TRAINING_DATA = []

# Create a Doc object for each text in TEXTS
for doc in nlp.pipe(TEXTS):
    # Match on the doc and create a list of matched spans
    spans = [doc[start:end] for match_id, start, end in matcher(doc)]
    # Get (start character, end character, label) tuples of matches
    entities = [(span.start_char, span.end_char, 'GADGET') for span in spans]
    
    # Format the matches as a (doc.text, entities) tuple
    training_example = (doc.text, {'entities': entities})
    # Append the example to the training data
    TRAINING_DATA.append(training_example)
    
print(*TRAINING_DATA, sep='\n')    
```

###**La boucle d’entraînement**

####**1. La boucle d’entraînement**
+ ***Alors que d’autres bibliothèques vous offrent une méthode qui prend en charge l’entraînement d’un modèle, spaCy vous donne un contrôle total sur la boucle d’apprentissage.***

####**2. Les étapes d’une boucle d’entraînement**

+ *La boucle d’apprentissage est une série d’étapes effectuées pour former ou mettre à jour un modèle. Nous devons généralement l’effectuer plusieurs fois, pour plusieurs itérations, afin que le modèle puisse en tirer des leçons efficacement. Si nous voulons nous entraîner pour 10 itérations, nous devons boucler 10 fois. Pour éviter que le modèle ne reste bloqué dans une solution sous-optimale, nous mélangeons aléatoirement les données pour chaque itération. C’est une stratégie très courante lors de la descente de gradient stochastique. Ensuite, nous divisons les données d’apprentissage en lots de plusieurs exemples, également connu sous le nom de minibatching. Cela facilite l’estimation plus précise du gradient. Enfin, nous mettons à jour le modèle pour chaque lot et recommençons la boucle jusqu’à ce que nous ayons atteint la dernière itération. Nous pouvons ensuite enregistrer le modèle dans un répertoire et l’utiliser dans spaCy.*

####**3. Récapitulatif : Fonctionnement de la formation**
+ Pour récapituler : 
  + **Les données d’apprentissage sont les exemples avec lesquels nous voulons mettre à jour le modèle**. 
  + **Le texte doit être une phrase, un paragraphe ou un document plus long**. 
  + **Pour obtenir les meilleurs résultats, il doit être similaire à ce que le modèle verra à l’exécution.** 
  + **L’étiquette est ce que nous voulons que le modèle prédise**. 
  + **Il peut s’agir d’une catégorie de texte ou d’une étendue d’entités et de son type**. 
  + **Le gradient est la façon dont nous devons modifier le modèle pour réduire l’erreur actuelle. Il est calculé lorsque nous comparons l’étiquette prédite à la vraie étiquette.**

####**4. Exemple de boucle**

Enfin, nous appelons la méthode nlp point sur disque pour enregistrer le modèle entraîné dans un répertoire.*

####**2. Problème 1 : Les modèles peuvent « oublier » des choses**


+ **Les modèles statistiques peuvent apprendre beaucoup de choses, mais cela ne signifie pas qu’ils ne les désapprendront pas. Si vous mettez à jour un modèle existant avec de nouvelles données, en particulier de nouvelles étiquettes, il peut surajuster et ajuster *trop* aux nouveaux exemples. Par exemple, si vous ne le mettez à jour qu’avec des exemples de « site Web », il peut « oublier » d’autres étiquettes qu’il avait précédemment prédites correctement – comme « personne ». C’est ce qu’on appelle aussi le problème de l’oubli catastrophique.**

####**3. Solution 1 : Intégrer des prédictions précédemment correctes**
+ **Pour éviter cela, assurez-vous de toujours mélanger des exemples de ce que le modèle a précédemment obtenu correctement. Si vous formez une nouvelle catégorie « site Web », incluez également des exemples de « personne ». spaCy peut vous y aider. Vous pouvez créer ces exemples supplémentaires en exécutant le modèle existant sur des données et en extrayant les étendues d’entités qui vous intéressent. Vous pouvez ensuite combiner ces exemples avec vos données existantes et mettre à jour le modèle avec des annotations de toutes les étiquettes.**

####**4. Problème 2 : Les modèles ne peuvent pas tout apprendre**

+ **Un autre problème courant est que votre modèle n’apprendra tout simplement pas ce que vous voulez. Les modèles de spaCy font des prédictions basées sur le contexte local – par exemple, pour les entités nommées, les mots environnants sont les plus importants. Si la décision est difficile à prendre en fonction du contexte, le modèle peut avoir du mal à l’apprendre. Le système d’étiquetage doit également être cohérent et pas trop spécifique. Par exemple, il peut être très difficile d’enseigner à un modèle si quelque chose est un vêtement pour adultes ou des vêtements pour enfants en fonction du contexte. Cependant, le simple fait de prédire l’étiquette « vêtements » peut mieux fonctionner.**

####**5. Solution 2 : planifiez soigneusement votre système d’étiquetage**

+ ***formation et la mise à jour des modèles, il vaut la peine de prendre du recul et de planifier votre système d’étiquettes. Essayez de choisir des catégories qui se reflètent dans le contexte local et de les rendre plus génériques si possible. Vous pouvez toujours ajouter un système basé sur des règles ultérieurement pour passer du générique au spécifique. Les catégories génériques comme « vêtements » ou « bande » sont à la fois plus faciles à étiqueter et plus faciles à apprendre.***
