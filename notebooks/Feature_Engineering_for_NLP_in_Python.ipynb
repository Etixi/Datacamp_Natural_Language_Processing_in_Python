####**Hashtags et mentions dans les tweets russes**

+ ***Revoyons la tweetstrame de données contenant les tweets russes. Dans cet exercice, vous allez calculer le nombre de hashtags et de mentions dans chaque tweet en définissant deux fonctions count_hashtags() et count_mentions() respectivement et en les appliquant à la fonctionnalité content de tweets.***

+ *Au cas où vous ne vous en souviendriez pas, les tweets sont contenus dans la fonction  content tweets.*

####**Consignes**

+ En compréhension de liste, permet startswith() de vérifier si un particulier wordcommence par '#'.

+ En compréhension de liste, permet startswith() de vérifier si un particulier wordcommence par '@'.

```python
# Function that returns numner of hashtags in a string
def count_hashtags(string):
	# Split the string into words
    words = string.split()

    # Create a list of words that are hashtags
    hashtags = [word for word in words if word.startswith('#')]

    # Return number of hashtags
    return(len(hashtags))
```

```python
# Create a feature hashtag_count and display distribution
tweets['hashtag_count'] = tweets['content'].apply(count_hashtags)
tweets['hashtag_count'].hist()
plt.title('Hashtag count distribution')
plt.show()
```

```python
# Function that returns number of mentions in a string
def count_mentions(string):
	# Split the string into words
    words = string.split()

    # Create a list of words that are mentions
    mentions = [word for word in words if word.startswith('@')]

    # Return number of mentions
    return(len(mentions))
```

```python
# Create a feature mention_count and display distribution
tweets['mention_count'] = tweets['content'].apply(count_mentions)
tweets['mention_count'].hist()
plt.title('Mention count distribution')
plt.show()
```

####**2. Aperçu des tests de lisibilité**

A noter que ces tests sont utilisés pour des textes en anglais. Il existe également des tests pour d'autres langues qui prennent en considération les nuances de cette langue particulière.***

####**5. Facilité de lecture Flesch**
+ ***Le Flesch Reading Ease est l'un des tests de lisibilité les plus anciens et les plus largement utilisés. Le score est basé sur deux idées : la première est que plus la longueur moyenne des phrases est grande, plus le texte est difficile à lire. Considérez ces deux phrases. Le premier est plus facile à suivre que le second. La seconde est que plus le nombre moyen de syllabes dans un mot est élevé, plus le texte est difficile à lire. Par conséquent, je vis dans ma maison est considéré comme plus facile à lire que je réside dans mon domicile en raison de son utilisation de syllabes inférieures par mot. Plus le score Flesch Reading Ease est élevé, plus la lisibilité est grande. Par conséquent, un score plus élevé indique que le texte est plus facile à comprendre.***

####**6. Interprétation des scores de facilité de lecture de Flesch**

+ ***Ce tableau montre comment interpréter les scores Flesch Reading Ease. Un score supérieur à 90 impliquerait que le texte est compréhensible pour un élève de 5e année, tandis qu'un score inférieur à 30 impliquerait que le texte ne peut être compris que par des diplômés universitaires.***

####**7. Indice de brouillard de tir**

+ ***L'indice de brouillard Gunning a été développé en 1954. Comme Flesch, ce score dépend également de la durée moyenne de la peine. Cependant, il utilise un pourcentage de mots complexes au lieu de syllabes moyennes par mot pour calculer son score. Ici, les mots complexes font référence à tous les mots qui ont trois syllabes ou plus. Contrairement à Flesch, la formule de l'indice de brouillard Gunning est telle que plus le score est élevé, plus le passage est difficile à comprendre.***

####**8. Interprétation de l'indice de brouillard de tir**

+ ***Ces fonctions prendront un morceau de texte et généreront une liste contenant les balises POS pour chaque mot. Il renverra alors le nombre de noms propres/autres noms que contient le texte. Nous utiliserons ces fonctions dans le prochain exercice pour générer des informations intéressantes sur les fausses nouvelles.***

+ *Le modèle en_core_web_sm a déjà été chargé comme nlp dans cet exercice.*

####**Consignes**

+ En utilisant la méthode count de la liste, comptez le nombre de noms propres (annotés par PROPN) dans la posliste.

+ En utilisant la méthode count de la liste, comptez le nombre d'autres noms (annotés par NOUN) dans la posliste.

```python
nlp = spacy.load('en_core_web_sm')

# Returns number of proper nouns
def proper_nouns(text, model=nlp):
  	# Create doc object
    doc = model(text)
    # Generate list of POS tags
    pos = [token.pos_ for token in doc]

    # Return number of proper nouns
    return pos.count('PROPN')

print(proper_nouns("Abdul, Bill and Cathy went to the market to buy apples.", nlp))
```

```python
nlp = spacy.load('en_core_web_sm')

# Returns number of other nouns
def nouns(text, model=nlp):
  	# Create doc object
    doc = model(text)
    # Generate list of POS tags
    pos = [token.pos_ for token in doc]

    # Return number of other nouns
    return pos.count('NOUN')

print(nouns("Abdul, Bill and Cathy went to the market to buy apples.", nlp))
```

####**2. modélisation n-gramme**

+ ***Dans la modélisation n-gramme, le poids d'une dimension pour la représentation vectorielle d'un document dépend du nombre de fois que le mot correspondant à la dimension apparaît dans le document. Disons que nous avons un document dans lequel le mot "humain" apparaît 5 fois. Alors, la dimension de sa représentation vectorielle correspondant à 'humain' aurait la valeur 5.***

####**3. Motivations**

+ ***Cependant, certains mots apparaissent très fréquemment dans tous les documents du corpus. En conséquence, les représentations vectorielles sont davantage caractérisées par ces dimensions. Considérons un corpus de documents sur l'Univers. Disons qu'il existe un document particulier sur Jupiter où les mots « jupiter » et « univers » apparaissent environ 20 fois. Cependant, « jupiter » figure rarement dans les autres documents alors que « univers » est tout aussi courant. Nous pourrions soutenir que bien que *jupiter* et *universe* se produisent 20 fois, *jupiter* devrait recevoir un poids plus important en raison de son exclusivité. En d'autres termes, le mot 'jupiter' caractérise le document plus que 'univers'.***

####**4. Candidatures**

+ ***La pondération des mots de cette manière a un grand nombre d'applications. Ils peuvent être utilisés pour détecter automatiquement les mots vides pour le corpus au lieu de s'appuyer sur une liste générique. Ils sont utilisés dans les algorithmes de recherche pour déterminer le classement des pages contenant la requête de recherche et dans les systèmes de recommandation, comme nous le découvrirons bientôt. Dans de nombreux cas, ce type de pondération génère également de meilleures performances lors de la modélisation prédictive.***

####**5. Fréquence du terme-fréquence inverse du document**

+ ***Le mécanisme de pondération que nous avons décrit est connu sous le nom de terme fréquence-fréquence de document inverse ou tf-idf en abrégé. Elle repose sur l'idée que le poids d'un terme dans un document doit être proportionnel à sa fréquence et inversement fonction du nombre de documents dans lesquels il apparaît.***

####**6. Formule mathématique**

+ ***Mathématiquement, le poids d'un terme i dans le document $j$ est calculé comme***

####**7. Formule mathématique**

+ ***terme fréquence du terme $i$ dans le document $j$***

####**8. Formule mathématique**

+ ***multiplié par le logarithme du rapport du nombre de documents dans le corpus et du nombre de documents dans lesquels le terme $i$ apparaît ou dfi.***

####**9. Formule mathématique**

+ ***Par conséquent, disons que le mot "bibliothèque" apparaît 5 fois dans un document. Il y a 20 documents dans le corpus et « bibliothèque » apparaît dans 8 d'entre eux. Ensuite, le poids tf-idf de 'bibliothèque' dans la représentation vectorielle de ce document sera de 5 fois log de 20 par 8, soit environ 2. En général, plus le poids tf-idf est élevé, plus le mot est important pour caractériser le document. Un poids tf-idf élevé pour un mot dans un document peut impliquer que le mot est relativement exclusif à ce document particulier ou que le mot apparaît extrêmement fréquemment dans le document, ou les deux.***

####**2. Formule mathématique**

+ ***En termes très simples, le score de similarité cosinus de deux vecteurs est le cosinus de l'angle entre les vecteurs. Mathématiquement, c'est le rapport du produit scalaire des vecteurs et du produit de la magnitude des deux vecteurs. Passons en revue ce que cette formule signifie vraiment.***

####**3. Le produit scalaire**

+ ***Le produit scalaire est calculé en additionnant le produit des valeurs sur les dimensions correspondantes des vecteurs. Disons que nous avons deux vecteurs à n dimensions $V$ et $W$ comme indiqué. Ensuite, le produit scalaire ici serait $v_1$ fois $w_1$ plus $v_2$ fois $w_2$ et ainsi de suite jusqu'à ce que $v_n$ fois $w_n$. A titre d'exemple, considérons deux vecteurs $A$ et $B$. En appliquant la formule ci-dessus, nous voyons que le produit scalaire revient à 37.***

####**4. Magnitude d'un vecteur**

+ ***La magnitude d'un vecteur est essentiellement la longueur du vecteur. Mathématiquement, il est défini comme la racine carrée de la somme des carrés des valeurs sur toutes les dimensions d'un vecteur. Par conséquent, pour un vecteur $V$ à $n$ dimensions, la magnitude, mod $V$, est calculée comme la racine carrée de $v_1$ carré plus $v_2$ carré et ainsi de suite jusqu'à vn carré. Considérez le vecteur A d'avant. En utilisant la formule ci-dessus, nous calculons sa magnitude comme étant la racine 66.***

####**5. Le score cosinus**

+ ***Nous sommes maintenant en mesure de calculer le score de similarité cosinus de $A$ et $B$. C'est le produit scalaire, qui est 37, divisé par le produit des grandeurs de $A$ et $B$, qui sont respectivement la racine 66 et la racine 38. La valeur ressort à environ 0,738, qui est la valeur du cosinus de l'angle thêta entre les deux vecteurs.***

####**6. Cosinus Score : points à retenir**

+ ***Étant donné que le score de cosinus est simplement le cosinus de l'angle entre deux vecteurs, sa valeur est comprise entre -1 et 1. Cependant, en NLP, les vecteurs de document utilisent presque toujours des poids non négatifs. Par conséquent, les scores de cosinus varient entre 0 et 1 où 0 indique aucune similitude et 1 indique que les documents sont identiques. Enfin, puisque le cosinus ne tient pas compte de l'amplitude des vecteurs, il est assez robuste pour documenter la longueur. Cela peut être un avantage ou un inconvénient selon le cas d'utilisation.***

Vous devez calculer la matrice de similarité cosinus qui contient le score de similarité cosinus par paires pour chaque paire de phrases (vectorisées à l'aide de tf-idf).***

+ *N'oubliez pas que la valeur correspondant à la ième ligne et à la jième colonne d'une matrice de similarité indique le score de similarité pour le ième et le jième vecteur.*

####**Instructions**

+ Initialiser une instance de TfidfVectorizer. Nommez-le `tfidf_vectorizer`.
+ À l'aide de `fit_transform()`, générez les vecteurs tf-idf pour corpus. Nommez-le tfidf_matrix.
+ Utilisez `cosine_similarity()` et passez `tfidf_matrix`pour calculer la matrice de similarité cosinus `cosine_sim`.

```python
corpus = ['The sun is the largest celestial body in the solar system',
          'The solar system consists of the sun and eight revolving planets',
          'Ra was the Egyptian Sun God',
          'The Pyramids were the pinnacle of Egyptian architecture',
          'The quick brown fox jumps over the lazy dog']
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Initialize an instance of tf-idf Vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Generate the tf-idf vectors for the corpus
tfidf_matrix = tfidf_vectorizer.fit_transform(corpus)

# Compute and print the cosine similarity matrix
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
print(cosine_sim)
```

####**2. Le problème avec BoW et tf-idf**

+ ***Considérez les trois phrases, je suis heureux, je suis joyeux et je suis triste. Maintenant, si nous devions calculer les similitudes, je suis heureux et je suis joyeux aurait le même score que je suis heureux et je suis triste, quelle que soit la façon dont nous le vectorisons. C'est parce que 'heureux', 'joyeux' et 'triste' sont considérés comme des mots complètement différents. Cependant, nous savons que la joie et la joie se ressemblent plus que la tristesse. C'est quelque chose que les techniques de vectorisation que nous avons couvertes jusqu'à présent ne peuvent tout simplement pas capturer.***

####**3. Incorporations de mots**

+ ***L'incorporation de mots est le processus de mappage de mots dans un espace vectoriel à n dimensions. Ces vecteurs sont généralement produits à l'aide de modèles d'apprentissage en profondeur et d'énormes quantités de données. Les techniques utilisées sortent du cadre de ce cours. Cependant, une fois générés, ces vecteurs peuvent être utilisés pour discerner à quel point deux mots sont similaires. Par conséquent, ils peuvent également être utilisés pour détecter des synonymes et des antonymes. Les incorporations de mots sont également capables de capturer des relations complexes. Par exemple, il peut être utilisé pour détecter que les mots roi et reine se rapportent de la même manière que l'homme et la femme. Ou que la France et Paris sont liés de la même manière que la Russie et Moscou. Une dernière chose à noter est que les incorporations de mots ne sont pas formées sur les données de l'utilisateur ; ils dépendent du modèle spatial pré-formé que vous utilisez et sont indépendants de la taille de votre jeu de données.***
