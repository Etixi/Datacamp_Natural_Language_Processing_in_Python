+ ***La fréquence est mesurée en kilohertz, mais est également appelée kHz ou taux d’échantillonnage***.

+ Avec un kilohertz égal à mille informations par seconde.

####**3. Exemples de fréquence**

###**Octets en nombres entiers**

+ Vous avez vu comment importer et lire un fichier audio à l'aide du module wave de Python et de la méthode **readframes()**. 

+ Mais cela se traduit par un tableau d'octets.

+ Pour convertir les octets en quelque chose de plus utile, nous utiliserons la méthode **frombuffer() de NumPy**.

+ En passant de **frombuffer()** nos octets d'ondes sonores et en indiquant un dtype de **'int16'**, nous pouvons convertir nos octets en entiers. 

+ Les entiers sont beaucoup plus faciles à utiliser que les octets.

+ *La bibliothèque Python wave a déjà été importée avec le fichier audio good_morning.wav.*

####**Des instructions**

+ Importez le package numpy avec son alias commun np.
+ Ouvrez et lisez le fichier audio bonjour.
+ Convertissez les octets signal_gm en entiers int16.
+ Afficher les 10 premières valeurs d'ondes sonores.

```python
import numpy as np

# Open good morning sound wave and read frames as bytes
good_morning = wave.open("/content/good-morning.wav", 'r')
signal_gm = good_morning.readframes(-1)

# Convert good morning audio bytes to integers
soundwave_gm = np.frombuffer(signal_gm , dtype='int16')

# View the first 10 sound wave values
print(soundwave_gm[:10])
```

###**Trouver les horodatages**
+ Nous savons que la fréquence de notre onde sonore est de 48 kHz, mais que se passerait-il si ce n'était pas le cas ? 

+ Nous pourrions le trouver en divisant la longueur de notre réseau d'ondes sonores par la durée de notre onde sonore. 

+ Cependant, le module wave de Python a un meilleur moyen. 

+ **L'appel de getframerate() sur un objet wave renvoie la fréquence d'images de cet objet wave.**

+ Nous pouvons ensuite utiliser la méthode **linspace() de NumPy** pour trouver l'horodatage de chaque entier dans notre tableau d'ondes sonores. 

+ Cela nous aidera à visualiser notre onde sonore dans le futur.

+ ***La méthode linspace() prend les paramètres start, stop et num et renvoie num valeurs régulièrement espacées entre start et stop.***

+ Dans notre cas, start sera égal à zéro, stop sera la longueur de notre réseau d'ondes sonores sur la fréquence d'images (ou la durée de notre fichier audio) et num sera la longueur de notre réseau d'ondes sonores.

####**Des instructions**

+ Convertissez les octets d'ondes sonores en nombres entiers.
+ Obtenez la fréquence d'images du fichier audio bonjour à l'aide de getframerate().
+ Définissez stop sur la longueur de soundwave_gm sur la fréquence d'images.
+ Définissez num comme étant la longueur de soundwave_gm.

```python
# Read in sound wave and convert from bytes to integers
good_morning = wave.open('/content/good-morning.wav', 'r')
signal_gm = good_morning.readframes(-1)
soundwave_gm = np.frombuffer(signal_gm , dtype='int16')

# Get the sound wave frame rate
framerate_gm = good_morning.getframerate()

# Find the sound wave timestamps
time_gm = np.linspace(start=0,
                      stop=len(soundwave_gm)/framerate_gm,
					  num=len(soundwave_gm))

# Print the first 10 timestamps
print(time_gm[:10])
```

```python
# Create function to convert audio file to wav
def convert_to_wav(filename):
  """Takes an audio file of non .wav format and converts to .wav"""
  # Import audio file
  audio = AudioSegment.from_file(filename)
  
  # Create new filename
  new_filename = filename.split(".")[0] + ".wav"
  
  # Export file as .wav
  audio.export(new_filename, format="wav")
  print(f"Converting {filename} to {new_filename}...")
 
# Test the function
convert_to_wav('/content/ex4_call_1_stereo_mp3.mp3')
```

```python
def show_pydub_stats(filename):
  """Returns different audio attributes related to an audio file."""
  # Create AudioSegment instance
  audio_segment = AudioSegment.from_file(filename)
  
  # Print audio attributes and return AudioSegment instance
  print(f"Channels: {audio_segment.channels}")
  print(f"Sample width: {audio_segment.sample_width}")
  print(f"Frame rate (sample rate): {audio_segment.frame_rate}")
  print(f"Frame width: {audio_segment.frame_width}")
  print(f"Length (ms): {len(audio_segment)}")
  return audio_segment
```

####**Des instructions**

+ Utilisez "reconnaître_google()" pour transcrire les données audio.

```python
# Importing the speech_recognition library
import speech_recognition as sr
```

```python
def transcribe_audio(filename):
  """Takes a .wav format audio file and transcribes it to text."""
  # Setup a recognizer instance
  recognizer = sr.Recognizer()
  
  # Import the audio file and convert to audio data
  audio_file = sr.AudioFile(filename)
  with audio_file as source:
    audio_data = recognizer.record(source)
  
  # Return the transcribed text
  return recognizer.recognize_google(audio_data)
```

####**Création d'une entité nommée personnalisée dans spaCy**


+ Si les entités nommées intégrées de spaCy ne suffisent pas, vous pouvez créer les vôtres en utilisant **la classe EntityRuler() de spaCy**.

+ ***EntityRuler() vous permet de créer vos propres entités à ajouter à un pipeline spaCy.***

+ *Vous commencez par créer une instance de EntityRuler() et lui passez le pipeline actuel, nlp.*

+ *Vous pouvez ensuite appeler add_patterns() sur l'instance et lui transmettre un dictionnaire du modèle de texte que vous souhaitez étiqueter avec une entité.*

+ *Une fois que vous avez configuré un modèle, vous pouvez l'ajouter au pipeline nlp en utilisant **add_pipe()**.*

+ Étant donné qu'Acme est une entreprise technologique, vous décidez de baliser le modèle "smartphone" avec la balise d'entité "PRODUCT".

+ spaCy a été importé et un document existe déjà contenant le texte transcrit de call_4_channel_2.wav.

####**Des instructions**

+ Importez EntityRuler depuis spacy.pipeline.
+ Ajoutez "smartphone" comme valeur pour la clé "motif".
+ Ajoutez l'instance EntityRuler(), règle, au pipeline nlp.
+ Imprimer les attributs d'entité contenus dans doc.

```python
import spacy
import re
from spacy.language import Language

nlp = spacy.load('en_core_web_sm')
boundary = re.compile('^[0-9]$')

@Language.component("component")
def custom_seg(doc):
    prev = doc[0].text
    length = len(doc)
    for index, token in enumerate(doc):
        if (token.text == '.' and boundary.match(prev) and index!=(length - 1)):
            doc[index+1].sent_start = False
        prev = token.text
    return doc
    
nlp.add_pipe("component", before='parser')
```

```python
# Import EntityRuler class
from spacy.pipeline import EntityRuler
# Check spaCy pipeline
print(nlp.pipeline)
```

```python
# Import EntityRuler class
from spacy.pipeline import EntityRuler

# Create EntityRuler instance
ruler = EntityRuler(nlp)

# Define pattern for new entity
ruler.add_patterns([{"label": "PRODUCT", "pattern": "smartphone"}])

# Update existing pipeline
nlp.add_pipe(ruler, before="ner")

# Test new entity
for entity in doc.ents:
  print(entity.text, entity.label_)
```

####**Créer un classificateur de texte en langue parlée**

+ Maintenant que vous avez transcrit certaines données audio d'appel client, nous allons créer un modèle pour classer si le texte de l'appel client est pre_purchase ou post_purchase.

+ Nous avons 45 exemples d'appels pré-achat et 57 exemples d'appels post-achat.

+ Les données sur lesquelles le modèle s'entraînera sont stockées dans train_df et les données sur lesquelles le modèle prédira sont stockées dans test_df.

+ Essayez d'imprimer le .head() de chacun d'entre eux sur la console.

+ *Nous allons construire un pipeline sklearn en utilisant CountVectorizer() et TfidfTransformer() pour convertir nos échantillons de texte en nombres, puis utiliser un classificateur MultinomialNB() pour savoir à quelle catégorie appartient chaque échantillon.*

+ *Ce modèle fonctionnera bien sur notre petit exemple ici, mais pour de plus grandes quantités de texte, vous voudrez peut-être envisager quelque chose de plus sophistiqué.*

####**Des instructions**
+ Créez text_classifier à l'aide de ***CountVectorizer(), TfidfTransformer() et MultinomialNB().***
+ Ajustez text_classifier sur train_df.text et train_df.label.
+ Créez une prédiction en appelant predict() sur text_classifier et en lui transmettant la colonne de texte de test_df.
+ Évaluez le modèle en comparant les prévisions au test_df.label.

```python
# Import text classification packages
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.model_selection import train_test_split
```

```python
# Split data into train and test sets
train_df.text, train_df.label, test_df.text, test_df.label = train_test_split(X=df["text"],y=df["label"],test_size=0.3)
```

```python
# Build the text_classifier as an sklearn pipeline
text_classifier = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('tfidf', TfidfTransformer()),
    ('classifier', MultinomialNB()),
])

# Fit the classifier pipeline on the training data
text_classifier.fit(train_df.text, train_df.label)
```

```python
# Evaluate the MultinomialNB model
predicted = text_classifier.predict(test_df.text)
accuracy = 100 * np.mean(predicted == test_df.label)
print(f'The model is {accuracy}% accurate')
```
